```python
import pandas as pd;
```

```python
bangaluru = pd.read_csv('/content/Bengaluru_House_Data.csv');
```

```python
print("Missing data before imputation:")
print(bangaluru.isnull().sum())
```

```python
most_frequent_location = bangaluru['location'].mode()[0]
bangaluru['location'].fillna(most_frequent_location, inplace=True)

# For 'size': Fill missing values with the most frequent value (mode)
most_frequent_size = bangaluru['size'].mode()[0]
bangaluru['size'].fillna(most_frequent_size, inplace=True)

# For 'society': Fill missing values with the most frequent value (mode)
most_frequent_society = bangaluru['society'].mode()[0]
bangaluru['society'].fillna(most_frequent_society, inplace=True)

# For 'bath': Fill missing values with the median (numerical column)
median_bath = bangaluru['bath'].mode()[0]
bangaluru['bath'].fillna(median_bath, inplace=True)

# For 'balcony': Fill missing values with the median (numerical column)
median_balcony = bangaluru['balcony'].mode()[0]
bangaluru['balcony'].fillna(median_balcony, inplace=True)
```

```python
print("\nMissing data after imputation:")
print(bangaluru.isnull().sum())
```
